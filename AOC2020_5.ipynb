```python
def get_seat(code):
    nrows = 128
    rows = [r for r in range(nrows)]
    for ch in code[:7]:
        nrows = int(nrows/2)
        #print(ch, nrows)
        if ch == 'F':
            rows = rows[:nrows]
        else:
            rows = rows[nrows:]
    
    ncols = 8
    cols = [r for r in range(ncols)]
    for ch in code[7:]:
        ncols = int(ncols/2)
        #print(ch, nrows)
        if ch == 'L':
            cols = cols[:ncols]
        else:
            cols = cols[ncols:]
    seatid = int(8* rows[0] + cols[0])
    return rows[0],cols[0],seatid
    
```

```python
with open(r'AOC2020_5.inp','r') as f:
    seats = [s.rstrip('\n') for s in f.readlines()]

seats
```

```python
ids = [get_seat(seat) for seat in seats]
```

```python
max([i[2] for i in ids])
```
