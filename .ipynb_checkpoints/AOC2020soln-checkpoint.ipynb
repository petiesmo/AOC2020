```python
#AOC2020 Day 2
with open('AOC2020_2.inp', 'r') as f:
    txt = f.readlines()
txt
txt1 = [abc.rstrip("\n") for abc in txt]
txt2 = [abc.replace("-"," ").replace(":","").replace(" ", ",") for abc in txt1]
txt3 = [t.split(",") for t in txt2]
txt4 = [abc for (low,high,char,abc) in txt3 if int(low) <= abc.count(char) <= int(high)]
len(txt4)
```

```python
#Part 2, start with txt3 with new criteria

txt5 = [abc for (p1,p2,char,abc) in txt3 if int(abc[int(p1)-1]==char) + int(abc[int(p2)-1]==char) == 1]
len(txt5)
```

```python
#AOC2020 Day 3
#Toboggan trees

with open('AOC2020_3.inp','r') as f3:
    arr = f3.readlines()
arr1 = [abc.rstrip('\n') for abc in arr]
arr1
```

```python
stepx,stepy = 3,1
posx,posy = 0,0
wstripe, hstripe = len(arr1[0]), len(arr1)

trees = 0
while posy < hstripe:
    #print(arr1[posy][posx])
    if arr1[posy][posx] == '#':
        trees += 1
    posx = (posx + stepx) % wstripe
    posy += stepy

print(trees)



```

```python
# Part 2
stepsx = [1,3,5,7,1]
stepsy = [1,1,1,1,2]
wstripe, hstripe = len(arr1[0]), len(arr1)

treelst = []
for stepx,stepy in zip(stepsx,stepsy):
    posx,posy = 0,0
    trees = 0
    while posy < hstripe:
        #print(arr1[posy][posx])
        if arr1[posy][posx] == '#':
            trees += 1
        posx = (posx + stepx) % wstripe
        posy += stepy
    treelst.append(trees)

print(treelst)
#Trees on each slope: [53, 282, 54, 54, 22]
```

```python
import functools, operator
product = functools.reduce(operator.mul, treelst)
product
#Answer: 958815792
```
