```python
#Part 1: Determine for each family ANY questions answered yes
with open(r'AOC2020_6.txt','r') as f:
    inp = f.readlines()
    
forms = [line.rstrip('\n') for line in inp]

families = []
family = []
for form in forms:
    if len(form) == 0:
        families.append(set(''.join(family)))
        #print(len(family))
        family = []
    else:
        family.append(form)
#Add final family
families.append(set(''.join(family)))
```

```python
ans = [len(fm) for fm in families]
sum(ans)
#Ans = 7027
```

```python
#Part 2: Determine for each family where ALL members answered yes
def getfamset(family):
    famset = set(family[0])
    for fam in family[1:]:
        famset = famset & set(fam)
    return famset

families = []
family = []
for form in forms:
    if len(form) == 0:
        families.append(getfamset(family))
        #print(len(family))
        family = []
    else:
        family.append(form)
#Add final family
families.append(getfamset(family))
families
```

```python
ans2 = [len(fm) for fm in families]
sum(ans2)
#Ans = 3579
```
