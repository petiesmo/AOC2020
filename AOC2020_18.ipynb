## Strange Math - New order of operations

Parethesis take precedence; '+' and '*' take equal precedence, left-to-right

Part 1: What is the sum of all the expressions? [My Ans: 13976444272545]

Part 2: 'Advanced': '+' now has higher precedence than '*'.  Now what is the sum? [My Ans: 88500956630893]

```
from pprint import pprint
#Load input deck
inputfile = r'/content/ACO2020_18.inp'
with open(inputfile,'r') as f:
  equations = f.readlines()
  expressions = [expr.rstrip('\n').split(' ') for expr in equations]
pprint(expressions[:5])
```

```
#Functions
def parse_exp(inarr):
  '''Interprets text string into expression.  Parens become lists'''
  validops = '*+'
  expr = [] #Receiver expression array
  temparr = inarr[:]  #Donor array
  lastoper = '+'
  while True:
    if not temparr:
      print('List is empty')
      return expr, False
    item = temparr.pop(0)
    if item in validops:
      expr.append(item)
    elif item.isnumeric():
      expr.append(int(item))
    elif item.startswith('('):
      temparr.insert(0,item[1:])
      inner_exp, temparr = parse_exp(temparr)
      expr.append(inner_exp)  #Appends the smaller expression into larger
    elif item == ')':
      return expr, temparr  #End of smaller expression
    elif item.endswith(')'):
      elems = item.split(')')
      expr.append(int(elems[0])) #Deliver the number to the current expression
      for _i in range(len(elems)-1):
        temparr.insert(0,')') #Add individual parens back to the donor array
    else:
      print(f'Do not know what {item} is')
      break
    #Debug: print(f'{exptotal} AND {expr} AND {temparr}')

def calculate(total, oper, num):
  total += int(num) if oper == '+' else 0
  total *= int(num) if oper == '*' else 1
  return total

def eval_expression1(expr):
  ''' Parens have highest precedence; Addition and Multiplication have equal precedence
  Evaluation moves left-to-right, nesting when parens reached.
  Refactoring for part 2: Extracting eval functions out of the Parse function
  '''
  tempexpr = []
  #Do parenthesis
  for term in expr:
      if type(term) is list:
          tempexpr.append(eval_expression1(term))
      else:
          tempexpr.append(term)
  print(tempexpr)
  #Do priority 1 (+ or *)
  t2 = tempexpr
  if len(t2) == 1:
      return t2[0]
  else:
      lastoper = '+'
      total = 0
      if all([t in '*+' for t in t2[1::2]]):
          for term in t2:
              if str(term) in '*+':
                lastoper = term
              else:
                total += int(term) if lastoper == '+' else 0
                total *= int(term) if lastoper == '*' else 1
      else:
        raise ValueError(f'Operators not recognized in {t2}')

  return total

def eval_expression2(expr):
  ''' Addition has precedence over Multiplication
  Extracting eval functions out of the Parse function
  '''
  tempexpr = []
  #Do parenthesis
  for term in expr:
      if type(term) is list:
          tempexpr.append(eval_expression2(term))
      else:
          tempexpr.append(term)
  print(tempexpr)
  #Do priority 1 (+)
  t1 = tempexpr
  t2 = []
  index = 0
  while index < len(t1):
      term = t1[index]
      if type(term) is int or term == '*':
          t2.append(term)
          index += 1
          continue
      elif term == '+':
          t2.append(t2.pop()+t1[index+1])
          index += 2
          continue
      else:
          raise ValueError(f'{term} at {index} not recognized')
  #Do priority 2 (*)
  if len(t2) == 1:
      return t2[0]
  else:
      total = 1
      if all([t=='*' for t in t2[1::2]]):
          for term in t2[::2]:
              total *= term
      else:
          raise ValueError(f'Something not right in {t2}')

  return total
```
